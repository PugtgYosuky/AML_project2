```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
import cv2
```

```python
street = os.path.join('dataset', 'seg_train', 'seg_train', 'street')
buildings = os.path.join('dataset', 'seg_train','seg_train',  'buildings')
mountains = os.path.join('dataset', 'seg_train','seg_train', 'mountain')
glaciers = os.path.join('dataset', 'seg_train','seg_train', 'glacier')
```

```python
street_images = os.listdir(street)
building_images = os.listdir(buildings)
mountains_images = os.listdir(mountains)
glaciers_images = os.listdir(glaciers)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(15, 7))
plt.axis('off')
# street
for i in range(2):
    image = cv2.imread(os.path.join(street, random.choice(street_images)))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axes[0, i].imshow(image)
    axes[0, i].set_title('Street')
    axes[0, i].axis('off')
# buildings
for i in range(2, 4):
    image = cv2.imread(os.path.join(buildings, random.choice(building_images)))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axes[0, i].imshow(image)
    axes[0, i].set_title('Building')
    axes[0, i].axis('off')
# Mountain
for i in range(2):
    image = cv2.imread(os.path.join(mountains, random.choice(mountains_images)))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axes[1, i].imshow(image)
    axes[1, i].set_title('Mountain')
    axes[1, i].axis('off')
# glacier
for i in range(2, 4):
    image = cv2.imread(os.path.join(glaciers, random.choice(glaciers_images)))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axes[1, i].imshow(image)
    axes[1, i].set_title('Glacier')
    axes[1, i].axis('off')
```
